Write a Python Program to find the Factorial of a Number?

```python
n = 5
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
factorial(n)
```

```python
n = 5
def factorial(n):
    if n == 0:
        return 1
        
    else:
        res = 1
        for i in range(2, n+1):
            res *= i
    return res
factorial(n)
```

#### Write a Python program to print the Fibonaaci sequence?
A: Fibonacci sequence is the integer sequence of 0, 1, 1, 2, 3, 5, 8.... The first two terms are 0 and 1. All other terms are obtained by adding the preceding two terms. This means to say the nth term is the sum of (n-1)th and (n-2)th term.

```python
num = 10
n1, n2 = 0,1
print("Fibonacci Series:", n1, n2, end=" ")
for i in range(2,num):
    n3 = n1 + n2
    n1 = n2
    n2 = n3
    print(n3, end=" ")
```

#### Write a Python program to check Armstrong Number?
A: Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.
eg. abcd... = an + bn + cn + dn + 

```python
#include <math.h>
#include <stdio.h>

# Python program to check if the number is an Armstrong number or not

# take input from the user
num = int(input("Enter a number: "))

# initialize sum
sum = 0

# find the sum of the cube of each digit
temp = num

while temp > 0:
   digit = temp % 10
   sum += digit ** 3
   temp //= 10
   print(temp)
   
   #print(digit)

# display the result
if num == sum:
       print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
    
```

#### Write a Python program to check Armstrong Number in an Interval?

```python
# Program to check Armstrong numbers in a certain interval

lower = 100
upper = 2000

for num in range(lower, upper + 1):

   # order of number
   order = len(str(num))
    
   # initialize sum
   sum = 0

   temp = num
   while temp > 0:
       digit = temp % 10
       sum += digit ** order
       temp //= 10

   if num == sum:
       print(num)
```

#### Write a Python program to find the sum of natural numbers?
